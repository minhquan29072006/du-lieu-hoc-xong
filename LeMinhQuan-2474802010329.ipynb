```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('punkt')
# nltk.download('stopwords')

# Văn bản đầu vào
text = "This is a simple example showing how to remove stopwords using NLTK."

# Tách từ
words = word_tokenize(text)

# Lấy danh sách stopword tiếng Anh
stop_words = set(stopwords.words('english'))

# Loại bỏ các stopword
filtered_words = [word for word in words if word.lower() not in stop_words]

# Hiển thị kết quả
print("Văn bản gốc:")
print(text)

print("\nSau khi loại bỏ stopword:")
print(filtered_words)
```

```
import nltk
from nltk.corpus import wordnet as wn

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('wordnet')
# nltk.download('omw-1.4')

# Nhập từ cần tra cứu
word = input("Nhập từ cần tìm: ")

# Lấy các synset của từ
synsets = wn.synsets(word)

if not synsets:
    print(f"Không tìm thấy từ '{word}' trong WordNet.")
else:
    print(f"\nCác định nghĩa và ví dụ của từ '{word}':\n")

    for i, syn in enumerate(synsets, start=1):
        print(f"Nghĩa {i}:")
        print("Định nghĩa:", syn.definition())

        examples = syn.examples()
        if examples:
            print("Ví dụ:")
            for ex in examples:
                print("-", ex)
        else:
            print("Ví dụ: Không có")

        print("-" * 50)
```

```
import nltk
from nltk.corpus import wordnet

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('wordnet')
# nltk.download('omw-1.4')

# Nhập từ cần tìm
word = input("Nhập từ cần tìm: ")

synonyms = set()
antonyms = set()

# Duyệt qua các nghĩa của từ
for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        # Thêm từ đồng nghĩa
        synonyms.add(lemma.name())

        # Thêm từ trái nghĩa (nếu có)
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

# Hiển thị kết quả
print("\nCác từ đồng nghĩa:")
if synonyms:
    for s in sorted(synonyms):
        print("-", s)
else:
    print("Không tìm thấy.")

print("\nCác từ trái nghĩa:")
if antonyms:
    for a in sorted(antonyms):
        print("-", a)
else:
    print("Không tìm thấy.")
```

```
import nltk
from nltk.corpus import wordnet as wn

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('wordnet')
# nltk.download('omw-1.4')

# Nhập hai danh từ
word1 = input("Nhập danh từ thứ nhất: ")
word2 = input("Nhập danh từ thứ hai: ")

# Lấy nghĩa đầu tiên của mỗi danh từ
synsets1 = wn.synsets(word1, pos=wn.NOUN)
synsets2 = wn.synsets(word2, pos=wn.NOUN)

if not synsets1:
    print(f"Không tìm thấy danh từ: {word1}")
elif not synsets2:
    print(f"Không tìm thấy danh từ: {word2}")
else:
    noun1 = synsets1[0]
    noun2 = synsets2[0]

    similarity = noun1.path_similarity(noun2)

    print(f"\nDanh từ 1: {noun1.name()}")
    print(f"Danh từ 2: {noun2.name()}")
    print(f"Độ giống nhau (Path Similarity): {similarity}")
```

```
import nltk
from nltk.corpus import wordnet as wn

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('wordnet')
# nltk.download('omw-1.4')

# Nhập hai động từ
verb1 = input("Nhập động từ thứ nhất: ")
verb2 = input("Nhập động từ thứ hai: ")

# Lấy các synset của hai động từ
synsets1 = wn.synsets(verb1, pos=wn.VERB)
synsets2 = wn.synsets(verb2, pos=wn.VERB)

if not synsets1:
    print(f"Không tìm thấy động từ '{verb1}'.")
elif not synsets2:
    print(f"Không tìm thấy động từ '{verb2}'.")
else:
    # Chọn nghĩa đầu tiên của mỗi động từ
    v1 = synsets1[0]
    v2 = synsets2[0]

    similarity = v1.path_similarity(v2)

    print("\nKết quả:")
    print("Động từ 1:", v1.name())
    print("Động từ 2:", v2.name())

    if similarity is not None:
        print("Độ giống nhau (Path Similarity):", similarity)
    else:
        print("Không thể tính được độ giống nhau.")
```

```
import nltk
import random
from nltk.corpus import names

# Chỉ cần chạy một lần nếu chưa tải dữ liệu
# nltk.download('names')

# Lấy danh sách tên
male_names = names.words('male.txt')
female_names = names.words('female.txt')

# Gắn nhãn cho dữ liệu
labeled_males = [(name, "male") for name in male_names]
labeled_females = [(name, "female") for name in female_names]

# Gộp hai tập dữ liệu
all_names = labeled_males + labeled_females

# Trộn ngẫu nhiên
random.shuffle(all_names)

# In 15 kết hợp đầu tiên
print("15 kết hợp ngẫu nhiên (tên, nhãn):\n")

for item in all_names[:15]:
    print(item)
```

```
import nltk
from nltk.corpus import names

# nltk.download('names')  # chạy lần đầu nếu chưa có

# Lấy danh sách tên
male_names = names.words('male.txt')
female_names = names.words('female.txt')

# Tạo dữ liệu có nhãn (name, gender)
labeled_names = [(name, 'male') for name in male_names] + \
                [(name, 'female') for name in female_names]

# Trích xuất ký tự cuối cùng + giữ nhãn
last_char_dataset = [
    (name[-1].lower(), gender)
    for (name, gender) in labeled_names
    if name  # đảm bảo không rỗng
]

# Hiển thị 20 kết quả đầu tiên
print("20 mẫu (ký tự cuối, nhãn):\n")
for item in last_char_dataset[:20]:
    print(item)
```
